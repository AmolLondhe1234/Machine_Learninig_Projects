```python
import pandas as pd
y={'f(x)':[2,10,6,8],"g(x)":[2,4,5,-5]}
x=[1,2,3,4]
```

```python
graph =pd.DataFrame(y,x)
```

```python
graph.plot(kind='line',grid=True,title="my graph",ylabel="my y title",xlabel="my x label")
```
